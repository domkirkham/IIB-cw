We first need to load the data, stored in "tennis_data.mat". The data consists of an array, $W$ containing the names of each player, and an array, G, containing the results of all of the matches in the season. 

```python
# set seed for reproducibility
# load data
data = sio.loadmat('./tennis_data.mat')
# Array containing the names of each player
W = data['W']
# loop over array to format more nicely
for i, player in enumerate(W):
    W[i] = player[0]
# Array of size num_games x 2. The first entry in each row is the winner of game i, the second is the loser
G = (data['G'] - 1).astype(int)
# Number of players
M = W.shape[0]
# Number of Games
N = G.shape[0]
```

# Gibbs sampling (This may take a minute)

```python
# number of iterations
num_iters = 1100
# perform gibbs sampling, skill samples is an num_players x num_samples array
skill_samples = []
for qqq in range(8):
    np.random.seed(qqq+1231)
    skill_samples.append(gibbs_sample(G, M, num_iters))#, random_nums)
```

```python
# We need to rotate all of our vectors by 45 degrees so that the line w1=w2 is mapped to the x axis. Then we can use the standard python CDF functions for square regions

def joint_skills(skills, players):
    
    probs = np.zeros((len(players), len(players)))
    for iind, i in enumerate(players):
        for jind, j in enumerate(players):
            if i != j:
                x1 = skills[i, :]
                x2 = skills[j, :]
                x1_rot = x1 * math.cos(math.pi/4) + x2 * math.sin(math.pi/4)
                x2_rot = -x1 * math.sin(math.pi/4) + x2 * math.cos(math.pi/4)
                joint_samples_rot = np.array([x1_rot, x2_rot])
                joint_mean = np.mean(joint_samples_rot, axis=1)
                joint_var = np.cov(joint_samples_rot, rowvar=1)
                rv = multivariate_normal(joint_mean, joint_var)
                probs[iind, jind] = rv.cdf([1000,0])
            else:
                probs[iind, jind] = 0.5
    return probs

def joint_outcomes(skills, players):
    
    probs = np.zeros((len(players), len(players)))
    for iind, i in enumerate(players):
        for jind, j in enumerate(players):
            if i != j:
                x1 = skills[i, :]
                x2 = skills[j, :]
                x1_rot = x1 * math.cos(math.pi/4) + x2 * math.sin(math.pi/4)
                x2_rot = -x1 * math.sin(math.pi/4) + x2 * math.cos(math.pi/4)
                joint_samples_rot = np.array([x1_rot, x2_rot])
                joint_mean = np.mean(joint_samples_rot, axis=1)
                joint_var = np.cov(joint_samples_rot, rowvar=1)
                joint_var += [[1,0],[0,1]]
                rv = multivariate_normal(joint_mean, joint_var)
                probs[iind, jind] = rv.cdf([1000,0])
            else:
                probs[iind, jind] = 0.5
    return probs

joint_samples = [skill_samples[1][15,:], skill_samples[1][0,:]]
joint_mean = np.mean(joint_samples, axis=1)
joint_var = np.cov(joint_samples, rowvar=1)
x, y = np.mgrid[-1:3:.01, -1:3:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
rv = multivariate_normal(joint_mean, joint_var)
plt.figure(figsize=(10,10))
plt.contourf(x, y, rv.pdf(pos))
plt.plot(np.linspace(-1, 3, 2), np.linspace(-1, 3, 2))
plt.xlabel(r'$w_{1}$')
plt.ylabel(r'$w_{2}$')
plt.ylim([-1,3])
plt.xlim([-1,3])

probs_joint = joint_skills(skill_samples[1], [15, 0, 4, 10])
outcomes_joint = joint_outcomes(skill_samples[1], [15, 0, 4, 10])
```

```python
def emp_skills(skills, players):
    
    probs = np.zeros((len(players), len(players)))
    for iind, i in enumerate(players):
        for jind, j in enumerate(players):
            if i != j:
                x1 = skills[i, :]
                x2 = skills[j, :]
                probs[iind, jind] = np.mean(x1 > x2)
            else:
                probs[iind, jind] = 0.5
    return probs

emp_probs = emp_skills(skill_samples[1], [15, 0, 4, 10])


fig, ax = plt.subplots(figsize=(15,10))
sns.set_context("poster")
sns.heatmap(emp_probs, cmap='RdBu', square=True, vmax=1, vmin=0, annot=True, lw=5, linecolor='grey', cbar=False, xticklabels=["Djokovic", "Nadal", "Federer", "Murray"], yticklabels=["Djokovic", "Nadal", "Federer", "Murray"])
ax.set_yticklabels(["Djokovic", "Nadal", "Federer", "Murray"], 
    rotation=0, va="center")

ax.xaxis.tick_top()
```

```python
def sorted_barplot(P, var, W):
    """
    Function for making a sorted bar plot based on values in P, and labelling the plot with the
    corresponding names
    :param P: An array of length num_players (107)
    :param W: Array containing names of each player
    :return: None
    """
    M = len(P)
    xx = np.linspace(0, M, M)
    plt.figure(figsize=(30, 30))
    sorted_indices = np.argsort(P)
    sorted_names = W[sorted_indices]
    sorted_errors = var[sorted_indices]
    plt.barh(xx, P[sorted_indices], xerr=sorted_errors, ecolor="grey")
    plt.yticks(np.linspace(0, M, M), labels=sorted_names[:, 0])
    plt.ylim([-2, 109])
    plt.xlim([0, 1.2])
    plt.show()
```

```python
# Empirical game win percentage, but normalised
unique_win, win_counts = np.unique(G[:,0], return_counts=True)
unique_loss, loss_counts = np.unique(G[:,1], return_counts=True)
no_wins = dict(zip(unique_win, win_counts))
no_losses = dict(zip(unique_loss, loss_counts))
for i in range(107):
    if i not in no_wins.keys():
        no_wins[i] = 0
    if i not in no_losses.keys():
        no_losses[i] = 0
```

```python
win_ratio = {k: no_wins[k] / float(no_wins[k] + no_losses[k]) for k in no_wins}
ratios = list(win_ratio.values())

ratio_zscore = (ratios - np.mean(ratios))/np.std(ratios)
```
